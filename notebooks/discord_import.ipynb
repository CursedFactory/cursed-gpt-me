# Discord Message Importer

This notebook demonstrates how to use the DiscordLoader to find and import Discord messages for GPT2 training.

```python
from pathlib import Path
from src.discord_loader import DiscordLoader
import json
import logging

from src.logging_config import get_logger

logger = get_logger(__name__)

# Initialize the DiscordLoader
discord_loader = DiscordLoader()

# Example: Find Discord messages in a directory
directory = "/Volumes/PortaOne/Datasets/discord_gdpr/Messages/c85338836384628736/messages.json"

# Find all messages.json files recursively
message_files = discord_loader.discord_find_messages(directory)

print(f"Found {len(message_files)} message files")
for file_path in message_files:
    print(f"  - {file_path}")
```

```python
# Load and inspect a Discord message file
if message_files:
    sample_file = message_files[0]
    
    with open(sample_file, 'r', encoding='utf-8') as f:
        messages = json.load(f)
    
    print(f"Loaded {len(messages)} messages from {sample_file}")
    
    # Show first few messages
    for i, msg in enumerate(messages[:3]):
        print(f"\nMessage {i+1}:")
        print(f"  Content: {msg.get('content', 'No content')[:100]}...")
        print(f"  Author: {msg.get('author', {}).get('username', 'Unknown')}")
        print(f"  Timestamp: {msg.get('timestamp', 'Unknown')}")
```

```python
# Process messages for GPT2 training
def extract_message_text(messages):
    """Extract text content from Discord messages."""
    texts = []
    for msg in messages:
        content = msg.get('content', '')
        if content:
            texts.append(content)
    return texts

if message_files:
    all_messages = []
    for file_path in message_files:
        with open(file_path, 'r', encoding='utf-8') as f:
            messages = json.load(f)
            all_messages.extend(messages)
    
    texts = extract_message_text(all_messages)
    print(f"Extracted {len(texts)} messages for training")
    
    # Show sample texts
    for i, text in enumerate(texts[:3]):
        print(f"\nSample {i+1}:")
        print(text[:200] + "..." if len(text) > 200 else text)
```
